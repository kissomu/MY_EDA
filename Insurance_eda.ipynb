## **Perform EDA on Insurance Data and identify the variables what are responsible for affecting the variable "charges"**

```python
df.nunique()
```

# Here we can see number of male and female children region wise.


```python
df.groupby(by = ["sex","region"])["children"].sum().reset_index()
```

```python
sns.heatmap(df.corr(),annot=True,cmap="Blues" ).set(title='Corelation')
```

# heatmap showing correlation, where highest correlation is 1(blue) and lowest  0.2(white)

```python
sns.barplot(x = 'smoker',
            y = 'charges',
            hue = 'region',
            data = df).set(title='Reason of affecting charges')
```

# conclusion - 1.According to this barplot we can see the reason of increasing charges is smokers.
# 2.Also area wise smokers and there charges
# 3.the one who smokes there charges are high as compare to the one who not smoke.
